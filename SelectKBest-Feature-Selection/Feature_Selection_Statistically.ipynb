# Feature Selection Methodology using Statistics:  
### Methods to try:  
1. ***Univariate Analysis*** where we try to **select those features that have the strongest relationship with the output variable.**  
####    - SelectKBest methodology:    
    - Correlation tells us about the linear relationship between the FEATURE TAG and the TARGET TAG, but in addition to check how much that linear rleationship is significant, we use **F-Test** statistic to evaluate the same. Here, the KPI is ***p_value*** which tells us about the significance of linear relationship between the feature and target.  
        - Lower the p_value, higher the significance as our Null Hypothesis is: **There is no significance relationship between the FEATURE and TARGET**. 
    - ***``SelectKBest``*** methodology provides us both, the <u><b>Correlation and F-test computation in one single go and return the highest contributing features statistically</b></u>.  

2. ***Information Gain***:  
    - Information gain calculates the reduction in entropy from the transformation of a dataset. It can be used for feature selection by evaluating the Information gain of each variable in the context of the target variable.

### SelectKBest Working:  

**``SelectKBest``** works differently.

*It takes as a parameter a score function, which must be applicable to a pair **(ùëã, ùë¶)**. The score function must return an array of scores, one for each feature ùëã[:,ùëñ] of ùëã (additionally, it can also return p-values, but these are neither needed nor required)*.   
    - ***SelectKBest then simply retains the first ùëò features of ùëã with the highest scores.***

So, for example, if we pass **chi2** as a score function, *SelectKBest will compute the **chi2 statistic between each feature of ùëã and ùë¶** (assumed to be class labels)*.   

   - A small value will mean the feature is independent of ùë¶.  
   - A large value will mean the feature is non-randomly related to ùë¶, and so likely to provide important information. Only ùëò features will be retained.


```python
class data_load:
    
    def load_data(feature_path,target_path):
        '''
        I/P:
            - feature_path,target_path: input string containing the location of the data file store locally
        
        Execution:
            - We load the data and print out the SHAPE of the Dataset in hand.
            - Return the NAME of the FEATURES in Hand
            
        RETURN:
            - dataframe: final dataframe which will go for further analysis
        '''
        feature_dataframe = pd.read_csv(feature_path).rename(columns={'Unnamed: 0':'timestamp'}).set_index('timestamp')
        feature_dataframe = feature_dataframe.loc[:, ~feature_dataframe.columns.str.endswith(".SV")]
        feature_dataframe = feature_dataframe.drop(feature_dataframe.filter(regex='X').columns, axis=1)
        #feature_dataframe = feature_dataframe.drop(feature_dataframe.filter(regex='Y').columns, axis=1)
        feature_dataframe = feature_dataframe.drop(feature_dataframe.filter(regex='Q').columns, axis=1)
        target_dataframe = pd.read_csv(target_path).rename(columns={'Unnamed: 0':'timestamp'}).set_index('timestamp')
        
        # creating duplicated dataframe
#         fd = feature_dataframe.copy()
#         feature_dataframe_copy = feature_dataframe.copy()
#         feature_dataframe_copy = feature_dataframe_copy.add_suffix('_copy')
#         feature_dataframe_copy = feature_dataframe_copy.mul(1.07)
        
#         feature_dataframe = pd.concat([feature_dataframe,feature_dataframe_copy],axis=1)
        
#         feature_dataframe_copy = fd.copy()
#         feature_dataframe_copy = feature_dataframe_copy.add_suffix('_copy_2')
#         feature_dataframe_copy = feature_dataframe_copy.mul(2.07)
 
#         feature_dataframe = pd.concat([feature_dataframe,feature_dataframe_copy],axis=1)
    
        print('Shape of the Input dataframe:',feature_dataframe.shape,'\n')
        print('Feature Columns in Hand:\n',feature_dataframe.columns.tolist())
        print()
        print('Target Feature in hand {}\n'.format(target_dataframe.columns.tolist()))
        
        return feature_dataframe,target_dataframe
    
class univariate_feature_selection:
    def univariate_methodology(feature_df,target_df,max_features_user_wants,scoring_function):
        
        '''
        Input:
            - feature_df: dataframe consisting features data
            - target_df: dataframe containing target data
            - max_features_user_wants: Maximum feature to be shortlisted from the CORRELATED pool of features
            
        Execution:
            - SelectKBest requires the dataset to be in Numpy characteristic.
            - Then we eventually specify the TOP features to choose and using which score_function
            - Then we return the final selected feature names which we can use to create a new model.
            
        Return:
            - Feature List Names 
            - Target Name
            - A Dataframe comprising Correlation Scores and P-Values for each feature
        '''
        start_time = datetime.now()
        features = feature_df.values
        target = target_df.values.ravel()
        # target = target.astype('int')
        # print('Feature shape:',features.shape)
        # print('Target shape:',target.shape)
        
        # feature extraction
        print('User has asked for maxx {} features'.format(max_features_user_wants))
        test = SelectKBest(score_func = scoring_function, k=max_features_user_wants)
        '''
        score_func: Function taking two arrays X and y, and returning a pair of arrays 
                    (scores, pvalues) or a single array with scores. 
        
        f_regression: F-value between label/feature for regression tasks.
        The goal of the F-test is to provide significance level. If you want to make sure the features your are 
        including 
        are significant with respect to your ùëù-value, you use an F-test. 
        If you just want to include the ùëò best features, you can use the correlation only.
        Ref: 
            - https://stats.stackexchange.com/questions/204141/difference-between-selecting-features-based-on-f-regression-and-based-on-r2
            - https://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.f_regression.html#sklearn.feature_selection.f_regression
            
        Alternatively we can use:
            - mutual_info_regression: Mutual information for a continuous target.
            Estimate mutual information for a continuous target variable.
            Mutual information (MI) between two random variables is a non-negative value, which measures the 
            dependency between the variables. It is equal to zero if and only if two random variables are independent,
            and higher values mean higher dependency.
            
            The function relies on nonparametric methods based on entropy estimation from k-nearest neighbors 
            distances.
        
        ref: https://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.SelectKBest.html#sklearn.feature_selection.SelectKBest
        '''
        
        fit = test.fit(features,target)
        feature_scores = fit.scores_[fit.get_support()]
        print('Completed SelectKBest function')

        mask = test.get_support()
        new_features = feature_df.columns[mask]
        prep_df = pd.DataFrame()
        prep_df['Columns'] = feature_df.columns
        prep_df['{}_scores'.format(scoring_function.__name__)] = fit.scores_
        prep_df['P_values'] = fit.pvalues_
        out_list = []
        for column in feature_df.columns:
            corr_tuple = pearsonr(feature_df[column], target)
            out_list.append(corr_tuple[0])
        
        prep_df['Correlation_scores'] = out_list
        # print(fit.scores_)
        print('Completed Seleting new features.')
        end_time = datetime.now()
        print('Total time taken for SelectKBest execution: {}'.format(end_time - start_time))
        return new_features,target,prep_df
    
    '''
    A Function to plot out the FEATURES selected after KBEST Algorithm.
    
    We show the below two points in the graph:
        - Pearson Correlation to show the linear relationship
        - p_value:
            - Null Hypothesis: There is no linear relationship between the FEATURE and TARGET
                - If p_value is greater than 0.05, we fail to reject the NULL HYPOTHESIS, so the features which are
                selected should have p_value <=0.05 to conclude the below:
                    - THERE IS LINEAR RELATIONSHIP BETWEEN THE FEATURE AND THE TARGET VARIABLE.
    '''
    def plot_join_plot(feature, target):
        g = sns.jointplot(x=feature, y=target, kind='reg', color='royalblue',height=5)
        # ax.annotate(stats.pearsonr)
        r, p = stats.pearsonr(feature, target)
        g.ax_joint.annotate(f'$\\rho = {r:.3f}, p = {p:.3f}$',
                            xy=(0.1, 0.9), xycoords='axes fraction',
                            ha='left', va='center',
                            bbox={'boxstyle': 'round', 'fc': 'powderblue', 'ec': 'navy'})
        g.ax_joint.scatter(feature, target)
        g.set_axis_labels(xlabel=feature.name, ylabel=target.name, size=12)

        
'''
Let's execute the CODE!
'''
lp = LineProfiler()
# lp_wrapper = lp(unv_fs.univariate_methodology)
# lp_wrapper(feature_df,target_df,max_features_user_wants,scoring_function)
# lp.print_stats()    

# Load and Data Description
ld = data_load
feature_file_path = 'Data/e5a33678-f312-413c-86ca-a75aeb8bed71_FeatureTag_TRAINING.csv' # Feature Tag file INPUT
target_file_path = 'Data/e5a33678-f312-413c-86ca-a75aeb8bed71_PredTarg_TRAINING.csv' # Target Tag file INPUT
feature_df,target_df = ld.load_data(feature_file_path,target_file_path)
target_feature_name = target_df.columns.tolist()[0]

# f_regression as Scoring Function
unv_fs = univariate_feature_selection
max_features_user_wants = 'all'
scoring_function = f_regression
print('Scoring function used for feature selection:-> {}'.format(scoring_function.__name__))
lp_wrapper = lp(unv_fs.univariate_methodology)
new_set_of_features,target_feature,f_reg_df_result = lp_wrapper(feature_df,target_df,max_features_user_wants,scoring_function)
#lp.print_stats()

#new_set_of_features,target_feature,f_reg_df_result = unv_fs.univariate_methodology(feature_df,target_df,max_features_user_wants,scoring_function)
print('The appropriate features shortlisted:\n',new_set_of_features.tolist())
print()

# f_reg_res_df = f_reg_df_result[f_reg_df_result.Columns.isin(new_set_of_features.tolist())]
# print(f_reg_res_df.head(max_features_user_wants))

# Executing Plots
# start_time = datetime.now()
# train_df = pd.concat([feature_df, target_df], axis=1)
# for col in new_set_of_features:
#     unv_fs.plot_join_plot(train_df[col], target_df[target_feature_name])
#     plt.show()
# end_time = datetime.now()
# print('Total time taken by the plots: {}'.format(end_time - start_time))
    
def print_hello():
    with open('f_reg_out.txt',"w") as f:
        sys.stdout = f
        print(lp.print_stats())
        
stdout_obj = sys.stdout             # store original stdout 
print_hello()
sys.stdout = stdout_obj  

# sys.stdout=open("f_reg_out.txt","w")
# print(lp.print_stats())
# sys.stdout.close()

# sys.stdout = open("f_reg_out.txt", "w")
# print(lp.print_stats())


# with open('f_reg_out.txt', 'w') as f:
#     print(lp.print_stats())
```

## SelectKBest Under the hood:  
   - The smaller the p-value, the more is the **significance of that relationship**.  
   - SelectKBest executes the **Correlation Score and Statistical Test p_value** between a FEATURE and the TARGET. After calculating the same, it sorts the features in ASCENDING order based upon the calculated **p-values**.  
       - Naively when we do it using **pearsonr** methodology a TWO-TAILED TEST is executed but ***SelectKBest*** provides us the option to use different kind of Statistical Test based upon our need.  
           - Example: **f_regression (regression tasks),chi2,mutual_info_classif (classification problem)**
